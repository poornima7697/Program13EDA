```python
q1=np.percentile(df['YearsExperience'],25,interpolation="midpoint")
q3=np.percentile(df['YearsExperience'],75,interpolation="midpoint")
iqr=q3-q1
upper=np.where(df['YearsExperience']>(q3+1.5*iqr))
print(upper)
lower=np.where(df['YearsExperience']<(q1-1.5*iqr))
print(type(lower))
```

```python
df.drop(upper[0],inplace=True)
df.drop(lower[0],inplace=True)
print(df.shape)
```
